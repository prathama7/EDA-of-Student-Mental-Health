As there seems to be a missing value in the dataset, we can manipulate it using Ostrich Theorem (i.e delete the row with missing value from the data set) or Mean Imputation.

```python
df['Do you have Depression?'] = df['Do you have Depression?'].map({'Yes':1 ,'No':0})
df['Do you have Anxiety?'] = df['Do you have Anxiety?'].map({'Yes':1 ,'No':0}) 
df['Do you have Panic attack?'] = df['Do you have Panic attack?'].map({'Yes':1 ,'No':0}) 
df['Did you seek any specialist for a treatment?'] = df['Did you seek any specialist for a treatment?'].map({'Yes':1 ,'No':0}) 

```

```python
df.head()
```

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

mental_health_by_course = df.groupby('What is your course?')[['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']].mean()


```

```python
# Plotting the data
plt.figure(figsize=(12, 8))
mental_health_by_course.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Proportion of Students with Mental Health Issues by Course')
plt.xlabel('Course')
plt.ylabel('Proportion')
plt.xticks(rotation=45, ha='right')
plt.legend(['Depression', 'Anxiety', 'Panic attack'], loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Display the data for clarity
mental_health_by_course
```

```python
# Analyze CGPA vs Depression
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='Do you have Depression?', y='What is your CGPA?', palette='Set2')
sns.swarmplot(data=df, x='Do you have Depression?', y='What is your CGPA?', color='.25')
plt.title('CGPA vs. Depression')
plt.xlabel('Depression (0: No, 1: Yes)')
plt.ylabel('What is your CGPA?')
plt.show()

```

```python
# Analyze CGPA vs Anxiety
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='Do you have Anxiety?', y='What is your CGPA?', palette='Set3')
sns.swarmplot(data=df, x='Do you have Anxiety?', y='What is your CGPA?', color='.25')
plt.title('What is your CGPA? vs. Anxiety')
plt.xlabel('Anxiety (0: No, 1: Yes)')
plt.ylabel('What is your CGPA?')
plt.show()

```

```python
# Analyze CGPA vs Panic attack
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='Do you have Panic attack?', y='What is your CGPA?', palette='Set1')
sns.swarmplot(data=df, x='Do you have Panic attack?', y='What is your CGPA?', color='.25')
plt.title('What is your CGPA? vs. Panic attack')
plt.xlabel('Panic attack (0: No, 1: Yes)')
plt.ylabel('What is your CGPA?')
plt.show()
```

```python
# Count the number of students seeking specialist treatment
specialist_treatment_count = df['Did you seek any specialist for a treatment?'].value_counts()

# Plotting the data
plt.figure(figsize=(8, 6))
sns.barplot(x=specialist_treatment_count.index, y=specialist_treatment_count.values, palette='pastel')
plt.title('Number of Students Seeking Specialist Treatment')
plt.xlabel('Sought Specialist Treatment (0: No, 1: Yes)')
plt.ylabel('Number of Students')
plt.show()

# Display the count data
specialist_treatment_count
```
